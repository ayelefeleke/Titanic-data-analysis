this is the data cleaning

```
# To Fill missing values
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].mode()[0])

# To Drop Cabin column only if it exists
if 'Cabin' in titanic.columns:
    titanic = titanic.drop(columns=['Cabin'])

#To Remove duplicates
titanic = titanic.drop_duplicates()
```

let check if it is cleaned

cabin is successfulyy removed and the dataset is cleaned

let analysed the data

```
survival_by_gender = titanic.groupby('Sex')['Survived'].mean()
print("Survival rate by gender:\n", survival_by_gender)
```

```
survival_by_class = titanic.groupby('Pclass')['Survived'].mean()
print("\nSurvival rate by class:\n", survival_by_class)
```

```
by_age_class = titanic.groupby('Pclass')['Age'].mean()
print("\nAverage age per class:\n", by_age_class)
```

```
# let Create age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Middle-Age', 'Senior']
titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=bins, labels=labels)

survival_by_agegroup = titanic.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival rate by age group:\n", survival_by_agegroup)
```

filtering

```
female_survived = titanic[(titanic['Sex'] == 'female') & (titanic['Survived'] == 1)]
display(female_survived)
```

```
children_survived = titanic[(titanic['Age'] <= 12) & (titanic['Survived'] == 1)]
display(children_survived)
```

```
first_class_survived = titanic[(titanic['Pclass'] == 1) & (titanic['Survived'] == 1)]
display(first_class_survived)
```
